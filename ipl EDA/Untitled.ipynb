```python
import pandas as pd

matches = pd.read_csv("matches[1].csv")
deliveries = pd.read_csv("deliveries[1].csv")
```

```python
matches.head()
```

```python
deliveries.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

matches["winner"].value_counts().head(10).plot(
    kind="bar",
    color="skyblue"
)

plt.title("Top 10 Winning Teams")
plt.xlabel("Teams")
plt.ylabel("Wins")
plt.xticks(rotation=45)

plt.show()
```

```python
top_batsmen = deliveries.groupby("batter")["batsman_runs"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
top_batsmen.plot(kind="bar", color="orange")

plt.title("Top 10 IPL Run Scorers")
plt.xlabel("Player")
plt.ylabel("Runs")
plt.xticks(rotation=45)

plt.show()
```

```python
top_bowlers = deliveries.groupby("bowler")["is_wicket"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
top_bowlers.plot(kind="bar", color="green")

plt.title("Top 10 IPL Wicket Takers")
plt.xlabel("Bowler")
plt.ylabel("Wickets")
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(12,6))

matches["venue"].value_counts().head(10).plot(
    kind="bar",
    color="purple"
)

plt.title("Top 10 IPL Venues")
plt.xlabel("Venue")
plt.ylabel("Number of Matches")
plt.xticks(rotation=90)

plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.countplot(data=matches, x="toss_decision")

plt.title("Toss Decision Analysis")
plt.xlabel("Decision")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(12,6))

matches["player_of_match"].value_counts().head(10).plot(
    kind="bar",
    color="red"
)

plt.title("Top 10 Player of the Match Winners")
plt.xlabel("Player")
plt.ylabel("Awards")
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(12,6))

matches["season"].value_counts().sort_index().plot(
    kind="line",
    marker="o"
)

plt.title("IPL Matches Played Each Season")
plt.xlabel("Season")
plt.ylabel("Number of Matches")

plt.grid(True)
plt.show()
```
